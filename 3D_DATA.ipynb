```python

# I am defining the shape of the 3D array here with time, latitude, and longitude
shape = (12, 12, 12)  #  The digits here is showing the number of time steps, latitude, longitude points I want create so you can adjust that, 
#SO I want to make it 12 months does why I made it 12,12,12.

# Now I am about to create a random values of temperature with normal temperature and I will convert it to Kelvin
temperature_data = np.random.uniform(20+273, 40+273, size=shape)
temperature_data
```

```python
# Define the coordinates for each dimension:
# Time: A range of 4 dates starting from '2025-01-01', with a daily frequency
#Now I told Python to range my date from 2025 and assign it to shape 1 of the shape =(4,4,4), check the one I did in Question 1, 
#remember python starts reading from 0 thats why I did periods=shape[0], so meaning the time will take the firstt column. You feel me?
time = pd.date_range('2025-01', periods=shape[0], freq='ME')
time
```

```python
# FOR THE SECOND COLUMN i WILL represent it by the lat as you said, and it will take shape 1 because python starts reading -
#from 0 and we gave the o column to time and and the 1 is for this lat, so i will generate random values for lat using linspace from -+10
#Latitude: 12 values linearly spaced from -10 to 10 which Ghana is within
lat = np.linspace(-10, 10, shape[1])
lat
```

```python
# This for the Longitudes:  spaced from -10 to 10
lon = np.linspace(-10, 10, shape[2])
lon
```

```python
## 3
```

```python
# specify the coordinates example the lat we did should show as lat, 
#so as lon and time, just as the ones we see in normal data, you see to select time in data to work with we do sel(time='2020'), 
#now we want to make this data as such to show the time, lat, lon so someone can cll those variables to work with.

#naming the dimension or column for time as time
time_coord = xr.DataArray(time, dims=["time"], name="time")

# specifying the dimension name lat as "lat"
lat_coord = xr.DataArray(lat, dims=["lat"], name="lat")

# this is for lon
lon_coord = xr.DataArray(lon, dims=["lon"], name="lon")
```

```python

# now I will be linking everything together, so we performed those task separately
# The temperature data is linked to time, latitude, and longitude through the `coords` argument

my_temperature = xr.DataArray(temperature_data,coords=[time_coord, lat_coord, lon_coord], dims=["time", "lat", "lon"], name="temperature")
my_temperature
```
