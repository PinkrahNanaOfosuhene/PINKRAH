# some trigonometry Plots

# (2,10,2) meaning it will start at 2 and end in 9 not 10 and difference of 2

```python
# x=np.arange(2,10)
```

```python
x = np.arange(0, 361, 5)
y =np.cos(np.deg2rad(x))
y1 =np.sin(np.deg2rad(x))
y2 = np.tan(np.deg2rad(x))

#figure size
plt.figure(figsize=(8,5))
plt.plot(x,y, label=('cos'), marker='o',markersize=5, color='red') #with linstyles you can use '--' or "-."
plt.plot(x,y1, label=('sin'), marker='*',markersize=5, color='yellow')
#plt.plot(x,y2, label=('tan'))

#labelling plot axis
plt.ylabel('y-axis', fontsize=24)
plt.xlabel('x-axis', fontsize=24)
plt.title('Trig function plot')

#limit of y
#plt.ylim=(-1,1)

#legends location
plt.legend(loc=1);
```

# Subplot/Panel plot

```python
fig = plt.figure(figsize=(12,8))

# spacing along width(left to right,)
plt.subplots_adjust(wspace=0.5, hspace=0.5)


# four box first box = 221, second =(222), third plot=(2,2,3)
ax=fig.add_subplot(2,2,1)
ax.plot(x,y)
ax.set_ylabel('cos')

ax1= fig.add_subplot(2,2,2)
ax1.plot(x,y1)
ax1.set_ylabel('sin')


ax2= fig.add_subplot(2,2,3)
ax2.plot(x,y2)
ax2.set_ylabel('tan');

#to set limit for sin because its plot its huge
ax2.set_ylim(-1,1)
```
