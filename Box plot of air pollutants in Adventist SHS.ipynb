```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Adventist = pd.read_csv('Adventist_SHS.csv')
selected_columns = Adventist[['temperature', 'pm10_0', 'pm02_5', 'pm01_0', 'timestamp',"humidity"]]
selected_columns['timestamp'] = pd.to_datetime(selected_columns['timestamp'])
selected_columns['month'] = selected_columns['timestamp'].dt.month

fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))
fig.suptitle('Box Plots of Selected Variables by Month', fontsize=16)
variables = ['temperature', 'pm10_0', 'pm02_5', 'pm01_0',"humidity"]
axs = axs.flatten()  
for i, var in enumerate(variables):
    sns.boxplot(x='month', y=var, data=selected_columns, ax=axs[i])
    axs[i].set_title(var)
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Value')
    axs[i].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show();
```
