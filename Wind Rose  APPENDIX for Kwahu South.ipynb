```python
# importing libriares 
import numpy as np
import matplotlib.pyplot as plt
from windrose import WindroseAxes
import pandas as pd
```

```python
# viewing the csv file with pandas
ds = pd.read_csv('POWER_Point_Daily_20230101_20231231_006d6585N_000d6654W_LST11.csv')
ds
```

```python
# viewing the headings 
ds.dtypes
```

```python

# Seting up the figure and axis
fig, ax = plt.subplots(subplot_kw={'projection': 'windrose'})

# Ploting wind rose
ax.bar(ds['WD10M'], ds['WS10M'], normed=True, opening=0.8, edgecolor='black', bins=np.arange(0, 6, 1), cmap=plt.cm.gist_ncar)

# Setting title and adjust font size
ax.set_title('Wind Rose for Kwahu South', fontweight='bold', fontsize=18, pad=20)

# Adjusting legend
ax.legend(title='Wind Speed (m/s)', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding a grid
ax.grid(True)

# Saving the image with a higher resolution
plt.savefig("Wind_Rose_assignment.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
